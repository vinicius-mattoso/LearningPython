#Carregando arquivo em formato excel

```python
excel=pd.read_excel('https://raw.githubusercontent.com/vinicius-mattoso/LearningPython/main/src/base_graficos_excel.xlsx')
excel.head()
```

# Como ja temos os nossos dados podemos fazer nossos gráficos

## Fazendo o plot com o matplotlib

https://matplotlib.org/3.5.0/tutorials/introductory/pyplot.html#:~:text=matplotlib.,the%20plot%20with%20labels%2C%20etc.

https://matplotlib.org/stable/gallery/

```python
# preparando os vetores
time=excel['time']
f_x=excel['f(x)']
y_x=excel['y(x)']
g_x=excel['g(x)']
```

```python
# red dashes, blue squares and green triangles
plt.plot(time, f_x, 'r--', time, y_x, 'bs', time, g_x, 'g^')
plt.show()
```

```python
# red dashes, blue squares and green triangles
plt.plot(time, f_x, 'r--', time, y_x, 'bs', time, g_x, 'g^')
plt.xlabel('time')
plt.ylabel('funtions values')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(time, f_x, label="f(t)")
ax.plot(time, y_x, label="y(t)")
ax.plot(time, g_x, label="g(t)")
ax.set(xlabel='time (s)', ylabel='functions values',
       title='My first python plot')
ax.grid()
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 16))
ax.plot(time, f_x, label="f(t)")
ax.plot(time, y_x, label="y(t)")
ax.plot(time, g_x, label="g(t)")
ax.set_title('My first python plot', fontsize=20)
ax.set_xlabel('Time (s)', fontsize=16)
ax.set_ylabel('functions values', fontsize=16)
leg = ax.legend(fontsize=16)
ax.grid()
# Set tick font size
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
	label.set_fontsize(16)

plt.show()
```

# Plotando com o plotly

```python
import plotly.express as px
df = excel
# função acumulativa
fig = px.ecdf(df, x="time",y="f(x)")
fig.show()
```

```python
import plotly.express as px
df = excel
fig = px.scatter(df, x="time",y="y(x)")
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(x=time, y=f_x, name='f(x)'))

fig.add_trace(
    go.Scatter(x=time, y=y_x, name='y(x)'))

fig.add_trace(
    go.Scatter(x=time, y=g_x, name='g(x)'))

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(x=time, y=f_x, name='f(x)'))

fig.add_trace(
    go.Scatter(x=time, y=y_x, name='y(x)'))

fig.add_trace(
    go.Scatter(x=time, y=g_x, name='g(x)'))
fig.update_layout(
    title="My first Dash plot in python",
    xaxis_title="Time(s)",
    yaxis_title="functions values",
    legend_title="functions names",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"
    ))

fig.show()
```
