# Bloco de Condicionais

IF teste inicial:

    BLOCO DE CODIGO INICIAL

ELIF Testes seguintes:

    BLOCO DE CODIGO 

ELSE:

    BLOCO DE CODIGO FINAL



No formato de cdigo:

```
if condio_1:
  executar ao 1
elif condio_2:
  execute ao 2
else:
  execute a ao 3
```

## If

  * O cdigo  s ir executar o bloco do IF quando a condio testada for verdadeira

````python
variavel = False


print('Estou aqui 1:')
if variavel == True:
  print('Condio verdadeira')

print('Estou aqui 2:')

print('\n')

print('Estou aqui 3:')
############################################################################
# OBS: Se no colocamos condicional nenhum ele entende if variavel == True #
############################################################################
if variavel:
  print('Condio verdadeira')

print('Estou aqui 4:')
````

````python
variavel = True

print('Estou aqui 1:')
if variavel:
  print('Condio verdadeira')

print('Estou aqui 2:')
````

````python
variavel = True

print('Estou aqui 1:')
##############################################################
# OBS: Para colocar a condio de "diferente" ou "not equal" #
##############################################################

if variavel != True:
  print('Condio verdadeira')

print('Estou aqui 2:')
````

## Else

  * O cdigo  s ir executar o bloco do Else em caso de falsa condio previamente testada

````python
nome_1 = 'Ana'
nome_2 = 'Joo'

if nome_1 == nome_2:
  print('So os mesmos nomes!!')

else:
  print('So nomes diferentes!')
````

````python
nome_1 = 'Pedro'
nome_2 = 'Pedro'

if nome_1 == nome_2:
  print('So os mesmos nomes!!')

else:
  print('So nomes diferentes!')
````

## Elif

  * Essas estrutura  utilizada caso queira avaliar mais de duas condies diferentes

````python
num = input('Fornea um nmero [ 0, 10] : ')
````

````python
# OBS: Tomar cuidado com o tipo de dados que estamos comparando
print(type(num))
````

````python
num = int(num)
````

````python
num
````

````python
if num == 0:
    print('Voc chutou o zero')

elif num >0  and num <= 5:
    print('Voc chutou um numero entre 1 e 5')

elif num>5  and num <= 9:
    print('Voc chutou um numero entre 6 e 9')

else:
 print('Voc chutou o 10')
````
