# Tipos "Containers"

Essa classifica√ß√£o de "Containers" √© s√≥ para facilitar no entendimento, uma vez que esses tipos de variaveis que ser√£o apresentadas na sequ√™ncia possuem um conjunto de variaveis do tipo "prim√°ria".

Por exemplo em uma lista, temos uma sequ√™ncia de palavras. Portanto a lista √© uma variavel que podemos classificar como "container".

## Vari√°veis de tipo **LISTA**

Variavel do tipo  Lista, contem uma sequ√™ncia mutavel dos tipos primitivos de vari√°veis.

```python
# Defini√ß√£o de uma lista a uma variavel

#uma lista de n√∫meros
X = [1, 2, 3, 4, 5]

#uma lista de palavras
Y = ['ana','Pedro',"Bruno"]

#uma lista de variaveis boleanas
Z = [True, False, True]
```

```python
# exibindo o valor armazenda na variavel X e Y
print(X)
print(Y)
print(Z)

# exibindo o tipo da variavel X e Y
print( type(X) )
print( type(Y) )
print( type(Z) )
```

### Pegando um elemento da lista


A variavel X √© do tipo list, mas e os termos contidos nela? 
Para conseguir identificar o tipo de cada elemento da lista precisamos avaliar sepadaramente esses termos.

```python
# Pegando o primeiro elemento da lista X
print(X[0])
print('')
# Pegando o ultimo elemento da lista X
print(X[-1])
```

```python
# Pegando alguns elemento da lista
print(X[0:2])
```

```python
print( type(X[0]) )

print( type(X[-1]) )

print( type(X[3]) )
```

## ADICIONANDO ELEMENTOS NA LISA DEPOIS DE CRIADA:

metodos: APPEND() | INSERT() | EXTEND()

### M√©todo .append()

Adicionando um elemento ao final da lista

```python
notas= [8, 5, 10, 2, 4]
print(notas)
```

```python
# Adicionando mais uma nota na lista notas
notas.append(9)

print(notas)
```

Se eu quiser colocar um elemento em uma determinada posi√ß√£o da lista?


### M√©todo .insert()

Adicionando elementos em uma determinada posi√ß√£o da lista

```python
casa=['sala','cozinha','banheiro']
valor = 'quarto'
posicao = 0
casa.insert(posicao, valor)
print(casa)
```

```python
casa=['sala','cozinha','banheiro']
valor = 'quarto'
posicao = 2
casa.insert(posicao, valor)
print(casa)
```

### M√©todo .extend()

Adicionando um conjunto elementos ao final de uma lista

```python
notas= [8, 5, 10, 2, 4]
print(notas)
```

```python
novas_notas=[10,5,7,6]
notas.extend(novas_notas)
print(notas)
```

O m√©todo .append() n√£o serve para esse caso n√£o? 

```python
notas= [8, 5, 10, 2, 4]
notas.append(novas_notas)
print(notas)
```

## REMOVENDO ELEMENTOS NA LISA DEPOIS DE CRIADA:

metodos: remove() | pop()

### M√©todo .remove()

Removendo um determinado elemento da lista.

```python
alunos = ['vinicius', 'pedro','lucas', 'matheus', 'ana', 'maria']
print(alunos)
```

```python
aluno_removido = 'vinicius'
alunos.remove(aluno_removido)
print(alunos)
```

Se mando remover uma aluna que n√£o estava na lista inicialmente, o que acontece?

```python
alunos = ['vinicius', 'pedro','lucas', 'matheus', 'ana', 'maria']
aluno_removido = 'luiza'
alunos.remove(aluno_removido)
```

### M√©todo .pop()

Removendo um determinado elemento da lista por meio do seu index.

```python
alunos = ['vinicius', 'pedro','lucas', 'matheus', 'ana', 'maria']
print(alunos)
```

```python
index = 1
alunos.pop(index)
```

```python
print(alunos)
```

Retorna o index do elemento, se houver, dentro de uma lista.

## Vari√°veis de tipo **Dicion√°rio**

Variavel do tipo  Dicion√°rio, contem uma chave e uma variavel guaradada por essa chave. Essa variavel guardada no dicion√°rio pode ser de qualquer tipo de variavel visto nesse c√≥diogo.


```python
meu_primeiro_dicionario={
    'chave_0':'Vinicius',
    'chave_1':10,
    'key_0':['vinicius','lucas','luiza'],
    'qqr_coisa':True,
    'test1':(1,2,3,4)
}
```

```python
print(meu_primeiro_dicionario)
```

### M√©todo para pegar as chaves do dicion√°rio

.keys()

```python
meu_primeiro_dicionario.keys()
```

### M√©todo para pegar os valores do dicion√°rio

.values()

```python
meu_primeiro_dicionario.values()
```

### M√©todo para pegar um valor do dicion√°rio por meio de uma chave

.get() 

```python
meu_primeiro_dicionario.get('key_0')
```

### M√©todo para pegar um valor do dicion√°rio por meio de uma chave

de maneira direta 

```python
meu_primeiro_dicionario['key_0']
```

### M√©todo para pegar um valor do dicion√°rio por meio de uma chave

.get() 

```python
meu_primeiro_dicionario.items()
```

```python
print(type(meu_primeiro_dicionario.items()))
```

```python
for key, value in meu_primeiro_dicionario.items():
    print(f'{key} : {value}')
```
